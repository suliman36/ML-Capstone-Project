### Count unique values in 'title' & 'address'

```python
address = []
count_address=0
  
# loop to count the unique values in title
for i in range(0, len(df['address'])):
    
    if df['address'][i] not in address: 
        
        address.append(df['address'][i])
          
        count_address += 1

print("No.of.Address :", count_address)
  
print("Address :", address)
```

```python
title = []
count_title=0
  
# loop to count the unique values in title
for i in range(0, len(df['title'])):
    
    if df['title'][i] not in title: 
        
        title.append(df['title'][i])
          
        count_title += 1

print("No.of.Titles :", count_title)
  
print("Titles :", title)
```

## EDA

#### Highest reserved Holiday Houses

```python
plt.figure(figsize=(15,10))

#plt.title('Holiday Homes Types With Highest Reservation' ,size =30)
fig = sns.countplot(x='title', data=df, palette='magma', order = df['title'].value_counts().index)
fig.set(xlabel='Holiday House', ylabel='Number of Reservation')
fig.set(title='Holiday Homes Types With Highest Reservation')
plt.xlabel("Holiday House" ,size=10)
plt.ylabel("Number of Reservation" ,size=10)
plt.show()
```

#### Holiday Houses which has ( Pool, Tent, Internet an Balcony )
##### These are only four features out of 26 features, any other feature could be ploted too

```python
fig, ax =plt.subplots(2,2, figsize=(15,15))

p1 = sns.countplot(df['pool'], palette = ['#67168c', '#cc9f18'], ax=ax[0,0])
p1.set(title='Pool')
p1.set_xlabel(' ')
p1.set_ylabel('Count')
p2 = sns.countplot(df['tent'], palette = ['#67168c', '#cc9f18'], ax=ax[0,1])
p2.set(title='Tent')
p2.set_xlabel(' ')
p2.set_ylabel('Count')
p3 = sns.countplot(df['internet'], palette = ['#67168c', '#cc9f18'], ax=ax[1,0])
p3.set(title='Internet')
p3.set_xlabel(' ')
p3.set_ylabel('Count')
p4 = sns.countplot(df['balcony'], palette = ['#cc9f18', '#67168c'], ax=ax[1,1])
p4.set(title='Balcony')
p4.set_xlabel(' ')
p4.set_ylabel('Count')

print('Holiday Houses Featuers')
```

#### TOP 5 address of Holdiay Houses with the highest reservation

```python
plt.figure(figsize=(15,10))
#plt.title('Top5 Holiday Houses Address With Highest Reservation',size =20)


fig = sns.countplot(x='address', data=df, palette='magma', order = df['address'].value_counts()[:5].index)
fig.set(title='TOP5 Adderess With High Reservations')
fig.set(xlabel='Address', ylabel='Number of Reservation')
plt.show(),
```

#### Price categories for each Holiday House

```python
CrosstabResult=pd.crosstab(index=df['pricing_category'],columns=df['title'])
print(CrosstabResult)
CrosstabResult.plot.bar(figsize=(15,10),
                        color=['#FCE2DB', '#80489C', '#FF8FB1', '#432C7A',
                               '#FAC213','#F77E21','#9F5F80','#7FB77E'],
                        title='Holiday Houses Price Categories',
                        xlabel='Price Category', ylabel='Number of Holiday Houses')

plt.legend(title="Holiday House")
```

#### Space categories for each Holiday House

```python
CrosstabResult2=pd.crosstab(index=df['spacing_category'],columns=df['title'])
print(CrosstabResult2)
CrosstabResult2.plot.bar(figsize=(15,10),color=['#FCE2DB', '#80489C', '#FF8FB1', '#432C7A',
                                                '#FAC213','#F77E21','#9F5F80','#7FB77E'],
                                                title='Holiday Houses Space Categories',
                                                xlabel='Space Category', ylabel='Number of Holiday Houses'
                        )


plt.legend(title="Holiday House")
```

#### Popularity for each Holiday House

```python
CrosstabResult3=pd.crosstab(index=df['Popularity'],columns=df['title'])
print(CrosstabResult3)
CrosstabResult3.plot.bar(figsize=(15,10), color=['#FCE2DB', '#80489C', '#FF8FB1', '#432C7A',
                                                 '#FAC213','#F77E21','#9F5F80','#7FB77E'],
                                                 title='Holiday Houses Popularity',
                                                 xlabel='Popularity', ylabel='Number of Holiday Houses'
                        )

plt.legend(title="Holiday House")
```
